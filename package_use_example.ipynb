```python
# Data generator

# This is just for the example, the data can follow any kind of model


import numpy as np
import matplotlib.pyplot as plt 

x = np.linspace(0, 1, 1000)

a0 = 4.02
a1 = 31.42
a2 = 11
a3 = 10
a4 = 11


y = [a0 + a1*i + a2 * np.sin(a3*i+a4) for i in x]
y = [i + np.random.normal(0,5) + np.exp(np.random.normal(0,1.3)) for i in y]
```

```python

#1 Import the package
from sigmclipfit.sgmfit import sgmfit

#2 Create function and initial values for the coefficients
i_val = [1, 10, 10, 10, 5] 
def fun(x, a0, a1, a2, a3, a4):
    return a0 + a1 * x + a2 * np.sin( a3 * x + a4)
# The function can be anything you want


#3 Set up the model with data and parameters
fit = sgmfit(x_data=x, y_data=y, model_function=fun, i_values=i_val, n_sigm=1.8)

#4 Make the fit with sigma clipping
fit.fit_model()         # By default:                fit.fit_model(keep_i_values = False, max_iter=np.inf)
                        # it can be changed like:    fit.fit_model(keep_i_values = True, max_iter=50)

#5 Draw the plot with the original data and data used for the fit
fit.fit_plot()

#6 Obtain results
fit.result
```
