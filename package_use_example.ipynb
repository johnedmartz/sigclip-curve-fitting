```python
# Data generator

import numpy as np
import matplotlib.pyplot as plt 

x = np.linspace(0, 1, 1000)

a0 = 4.02
a1 = 31.42
a2 = 11
a3 = 10
a4 = 11

y = [a0 + a1*i + a2 * np.sin(a3*i+a4) for i in x]
y = [i + np.random.normal(0,5) + np.exp(np.random.normal(0,1.3)) for i in y]
```

```python
from sigmclipfit.sgmfit import sgmfit

i_val = [1, 10, 10, 10, 5] 
def fun(x, a0, a1, a2, a3, a4):
    return a0 + a1 * x + a2 * np.sin( a3 * x + a4)


fit = sgmfit(x, y, fun, i_val, 1.8)
fit.fit_model()
fit.fit_plot()

fit.result
```
